```python
plt.scatter(db1.TV,db1.Sales)
plt.title("Sales vs TV advertising")
plt.xlabel("TV advertising")
plt.ylabel("Sales")
plt.show()
```

## 5 ols

```python
from statsmodels.formula.api import ols
```

```python
reg_TV=ols("Sales ~ TV", data=db1).fit()
print(reg_TV.summary())
```

If you increase the TV advertising investing by $1.00

then sales will increase by $0.05

```python
reg_Radio=ols("Sales ~ Radio", data=db1).fit()
print(reg_Radio.summary())
```

```python
reg_Newspaper=ols("Sales ~ Newspaper", data=db1).fit()
print(reg_Newspaper.summary())
```

```python
reg_Total=ols("Sales ~ TV + Radio + Newspaper", data=db1).fit()
print(reg_Total.summary())
```

# Base 2

## 2 - Database

```python
db2=pd.read_csv("kc_house_data.csv")
db2
```

## 3 - Analysis

```python
plt.scatter(db2.bedrooms, db2.price)
plt.title("Prive vs Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.show()
```

```python
reg_bedrooms=ols("price ~ bedrooms", data=db2).fit()
print(reg_bedrooms.summary())
```

```python
reg_bathrooms=ols("price ~ bathrooms", data=db2).fit()
print(reg_bathrooms.summary())
```

```python
reg_sqft_living=ols("price ~ sqft_living", data=db2).fit()
print(reg_sqft_living.summary())
```

```python
reg_sqft_lot=ols("price ~ sqft_lot", data=db2).fit()
print(reg_sqft_lot.summary())
```

```python
reg_floors=ols("price ~ floors", data=db2).fit()
print(reg_floors.summary())
```

```python
reg_zipcode=ols("price ~ zipcode", data=db2).fit()
print(reg_zipcode.summary())
```

```python
reg_total=ols("price ~ bedrooms + bathrooms + sqft_living + sqft_lot + floors + waterfront + view + condition + grade + sqft_basement + yr_built + yr_renovated + zipcode", data=db2).fit()
print(reg_total.summary())
```

```python
reg_GLL=ols("price ~ grade + lat + long", data=db2).fit()
print(reg_GLL.summary())
```

```python
reg_total=ols("price ~ bedrooms + bathrooms + sqft_living + sqft_lot + floors + waterfront + view + condition + grade + yr_built + lat + long + sqft_living15 + sqft_lot15", data=db2).fit()
print(reg_total.summary())
```

```python
reg_total=ols("price ~ bedrooms + sqft_living + floors + waterfront + view + condition + grade + lat + long", data=db2).fit()
print(reg_total.summary())
```

These are the variables that affect directly on a house, on a positive way we got the square foot of living, waterfront, view, condition, grade and lattitude; on a negative way the variables are the number of bedrooms, floors and longitude
