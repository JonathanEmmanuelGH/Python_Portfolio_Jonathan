BASIC ALGORITHM OF MACHINE LEARNING

```python
import pandas as pd
```

```python
db=pd.read_csv("Student_Marks.csv")
```

```python
db
```

We import the functions from sklearn to apply machine learning

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

y is the dependent variable and X is the matrix for the independent variables

```python
y=db["Marks"]
```

```python
X=db[["number_courses","time_study"]]
```

Let's split the database between test and train

```python
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

ALGORITHM

1. Create the object

```python
reg=LinearRegression()
```

2. Adjust the object to the training part of the database (X_train and y_train)

```python
reg.fit(X_train,y_train)
```

3. The coefficients

```python
reg.intercept_
```

```python
reg.coef_
```

These are the coefficients of the regression interpretetion: 1.82 means if we increase 1 more course then my grade will increase in 1.82 points; 5.37 mean if increase 1 more hour of studying then my grade will increase in 5.37 points

4. I evaluate the model with the training part

```python
reg.score(X_train,y_train)
```

5. Predict with the test part

```python
y_pred=reg.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
error=y_pred-y_test
```

```python
error
```

```python
error.mean()
```

```python
import pandas as pd
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import train_test_split
```

```python
db2=pd.read_csv("advertisingp.csv")
```

```python
y=db2["Sales"]
X=db2[["TV","Radio","Newspaper"]]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
reg2=ElasticNet()
```

```python
reg2.fit(X_train,y_train)
```

```python
reg2.intercept_
```

```python
reg2.coef_
```

```python
reg2.score(X_train,y_train)
```

```python
y_pred=reg2.predict(X_test)
```

```python
error=y_pred-y_test
```

```python
error.mean()
```
