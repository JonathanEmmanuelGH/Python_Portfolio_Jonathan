```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
import numpy as np
```

```python
db = pd.read_csv("Student_Marks.csv")
```

```python
db
```

# 1

```python
plt.scatter(db.number_courses, db.Marks)
plt.title("Marks vs Number courses")
plt.xlabel("Number Courses")
plt.ylabel("Marks")
plt.show()
```

As we can see on the graph ther is no strong or lineal correlation, there's no a clear visualization of the model so the results on the students won't be tottally affected by the number of courses they got

```python
plt.scatter(db.time_study, db.Marks)
plt.title("Marks vs Number courses")
plt.xlabel("Time Study")
plt.ylabel("Marks")
plt.show()
```

This is a strong, positive correlational model in which we can see that Marks are strongly dependent on time study dedicated

# 2

```python
modelo_1 = ols("Marks ~ number_courses + time_study", data=db).fit()
print(modelo_1.summary())
```

Explaining 94% of the database we get that for each unit spend on studying a student might be able to increase their marks by 5.3 points and on the ohter hand we got that for each courses the students take they would increase their marks and grades by 1.8 points.

# 3

```python
modelo_2 = ols("np.log(Marks) ~ np.log(number_courses) + np.log(time_study)", data=db).fit()
print(modelo_2.summary())
```

With a R squared equal to 0.47% we got that for each percentage spent on studying (time_study) students should be able to increase their reasults by 47%. Then we got that increasing the number of courses, their marks would increase by 0.6%.

```python
modelo_3 = ols("Marks ~ np.log(number_courses) + np.log(time_study)", data=db).fit()
print(modelo_3.summary())
```

Using the 63.1% of our database, we got tthat increasing their time studying by 1% their grades would aumentate by 0.0971 points. After that, we got that increasing their number of courses by 1% we got an increased on their grades by 0.1246 points

```python
modelo_4 = ols("np.log(Marks) ~ number_courses + time_study", data=db).fit()
print(modelo_4.summary())
```

Using the 97.2% of our database, we got that increasing their time studying by 1 unit their grades would aumentate by 23.8%. After that, we got that increasing their number of courses by 1 unit we got an increased on their grades by 9.32%
