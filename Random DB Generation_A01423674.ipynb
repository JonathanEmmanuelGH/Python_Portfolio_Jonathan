```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Configuración inicial
start_date = datetime(2023, 10, 8)
end_date = datetime(2023, 11, 5)
date_range = (end_date - start_date).days
activities = [
    "Asistir a un seminario", "Comprar libros", "Ir al cine", "Cena en restaurante",
    "Viaje de fin de semana", "Concierto de música", "Partido de fútbol", "Sesión de gimnasio",
    "Curso de cocina", "Clases de baile", "Voluntariado", "Práctica de instrumento musical",
    "Estudio en grupo", "Compra de ropa", "Visita a museo", "Café con amigos",
    "Asistencia a obra de teatro", "Participación en hackathon", "Sesión de yoga", "Juegos de mesa con amigos",
    # Añade aquí más actividades según sea necesario
]
types = {
    "Alimentos/Salud": 1,
    "Ahorro/Inversión": 2,
    "Ejercicio/Deporte": 3,
    "Entretenimiento/Ocio": 4,
    "Académico": 5,
    "Transporte": 6
}
moments = {1: "mañana", 2: "tarde", 3: "noche"}

# Generar datos aleatorios
data = []
for _ in range(100):
    date = start_date + timedelta(days=random.randint(0, date_range))
    activity = random.choice(activities)
    cost = round(random.uniform(10, 500), 2)  # Costo entre 10 y 500 con dos decimales
    budget = round(cost + random.uniform(-30, 30), 2)  # Presupuesto en un rango de +-30 del costo
    time_spent = random.randint(30, 180)  # Tiempo invertido entre 30 y 180 minutos
    activity_type = random.choice(list(types.values()))  # Tipo de actividad
    moment = random.choice(list(moments.keys()))  # Momento del día
    number_of_people = random.randint(1, 5)  # Número de personas involucradas entre 1 y 5

    data.append([
        date.strftime("%d/%m/%Y"), activity, cost, budget, time_spent,
        activity_type, moments[moment], number_of_people
    ])

# Crear DataFrame
df = pd.DataFrame(data, columns=[
    "Fecha", "Nombre Actividad", "Costo", "Presupuesto",
    "Tiempo Invertido", "Tipo", "Momento", "No de Personas"
])

# Guardar en CSV
df.to_csv("base_de_datos.csv", index=False)

print("Base de datos generada y guardada en 'base_de_datos.csv'.")
```
