[Box Plot](https://pandas.pydata.org/docs/reference/api/pandas.plotting.boxplot.html)

```python
db.boxplot(column="math")
```

In the box the information is organized in quartiles (4 different parts) and the lines are the whiskers the smallest and the highest numbers. 5 is the lowest grade in math. 10 is the highest grade in math. 6 is the first quartile (if you get 6 you are above 25% of the grades) 7 is the second quartile (if you get 7 you are above the 50%) 8 is the third quartile (if you get 7 you are above 75% of the population and you are at the top 25%)

[Histogram](https://pandas.pydata.org/docs/reference/api/pandas.Series.plot.hist.html#pandas.Series.plot.hist)

```python
db["statistics"].hist()
```

[Matplotlib](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axvline.html://) (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9,4))

plt.hist(db["math"],bins=10,color="cornflowerblue")

plt.axvline(db["math"].mean(),color="red",label="Mean")
plt.axvline(db["math"].median(),color="yellow",label="Median")
plt.axvline(db["math"].mode()[0],color="green",label="Mode")

plt.xlabel("Math")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
