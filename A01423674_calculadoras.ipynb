#Calculadora de ahorro

```python
print("|°°°°°Bienvenido a esta calculadora de ahorro°°°°°|")
```

```python
print("Nombre: ")
nombre = input()
```

```python
Saldo_ini = float(input("Saldo inicial: "))
saldo_final = float(input("Saldo final: "))
meses = int(input("Número de mensualidades esperedas: "))
```

```python
ahorro = (saldo_final - Saldo_ini)/meses
```

```python
print(f"{nombre}, Si quieres tiener un ahorro final de ${round(saldo_final, 0)} debes de ahorrar ${round(ahorro)} durante {meses} meses")
```

#Calculadora de interés

```python
print("|°°°°°Bienvenido a esta calculadora de interés°°°°°|")
```

```python
tasa_inter = 0.08
interes_acumulado = 0
nombre_interes = input("Nombre: ")
saldo = int(input("Saldo inicial: "))
deposito = int(input("Cantidad a depositar en Octubre "))
retiro = int(input("Cantidad a retirar en Diciembre: "))
```

```python
for mes in range (1, 12 + 1):
  inter_mensual = saldo + tasa_inter / 12
  interes_acumulado = interes_acumulado + inter_mensual
  saldo = saldo + deposito
  if mes == 10:
    saldo = saldo + deposito
  if mes == 12:
    saldo = saldo - retiro
  print(f"El saldo en el mes {mes} es ${round(saldo, 0)}")
```

```python
print(f"El saldo al final de año en la cuenta de inversión será: ${round(saldo, 0)}")
print(f"El interés ganado en un año será: ${round(interes_acumulado, 0)}")
```
